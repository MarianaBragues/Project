Projeto: Search EC numbers on the excel file from uniprot 

```python
import pandas as pd
import re

def search_ec_numbers(input_file, ec_numbers, output_file):
    # Load the Excel file
    df = pd.read_excel(input_file)

    # Search for exact EC numbers in all columns
    result = pd.DataFrame()
    for col in df.columns:
        for ec in ec_numbers:
            pattern = r'\b{}\b'.format(ec)
            matches = df[df[col].astype(str).str.contains(pattern, flags=re.IGNORECASE, regex=True)]
            result = pd.concat([result, matches])

    # Remove duplicates from the results
    result = result.drop_duplicates()

    # Save the results to another Excel file
    result.to_excel(output_file, index=False)
```
